```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import  DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('data set.csv')
```

```python
df.isnull().sum()
```

```python
# to check percentage of missing values
(df['1/1/2014'].isna().sum()/len(df['1/1/2014']))*100
```

```python
#filling mean value in place of null values
df['1/1/2014'].fillna(df['1/1/2014'].mean())
```

```python
df.head()
```

```python
# Show number of NaN values before replacing
print("NaN values before replacement:\n", df.isna().sum())
```

```python
# Replace NaN with column-wise mean
df_filled = df.fillna(df.mean(numeric_only=True))
```

```python
# Show number of NaN values after replacement
print("\nNaN values after replacement:\n", df_filled.isna().sum())
```

```python
df_filled.head(10)
```

```python
df_filled.isnull().sum()
```

```python
df_filled.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_filled["CONS_NO"] = le.fit_transform(df_filled["CONS_NO"])
```

```python
print(df_filled.dtypes["CONS_NO"])
```

```python
df_filled.head(10)
```

```python
# split the data into training ans testing set
x=df_filled.drop('FLAG',axis=1)
y=df_filled['FLAG']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# train a random forest classifier
model=RandomForestClassifier(random_state=42)
```

```python
model.fit(x_train,y_train)
# make predictions
y_pred=model.predict(x_test)
```

```python
#evaluate the model
accuracy=accuracy_score(y_test,y_pred)
print(f'Accuracy: {accuracy:.2f}')
print('\nClassification Report:\n',classification_report(y_test,y_pred))
```
